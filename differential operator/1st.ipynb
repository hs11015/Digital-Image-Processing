```python
def get_Px_filter(filter_size):
    filter = np.zeros(shape=(filter_size, filter_size))
    for y in range(filter_size):
        filter[y][0] = -1
        filter[y][1] = 0
        filter[y][2] = 1
    return filter


def Px_filtering(src, filter_size):
    Px = np.zeros(shape=(src.shape))
    Px_filter = get_Px_filter(filter_size)
    #print(Px_filter)
    for y in range(int(filter_size/2), src.shape[0]-int(filter_size/2)):
        for x in range(int(filter_size/2), src.shape[1]-int(filter_size/2)):
            for j in range(filter_size):
                for i in range(filter_size):
                    Px[y][x] += src[y-int(filter_size/2)+j][x-int(filter_size/2)+i] * Px_filter[j][i]
    Px = np.array(np.round(Px), dtype=np.uint8)
    return Px
    
```

```python
def get_Py_filter(filter_size):
    filter = np.zeros(shape=(filter_size, filter_size))
    for x in range(filter_size):
        filter[0][x] = 1
        filter[1][x] = 0
        filter[2][x] = -1
    return filter


def Py_filtering(src, filter_size):
    Py = np.zeros(shape=(src.shape))
    Py_filter = get_Py_filter(filter_size)
    #print(Py_filter)
    for y in range(int(filter_size/2), src.shape[0]-int(filter_size/2)):
        for x in range(int(filter_size/2), src.shape[1]-int(filter_size/2)):
            for j in range(filter_size):
                for i in range(filter_size):
                    Py[y][x] += src[y-int(filter_size/2)+j][x-int(filter_size/2)+i] * Py_filter[j][i]
    Py = np.array(np.round(Py), dtype=np.uint8)
    return Py
    
```
