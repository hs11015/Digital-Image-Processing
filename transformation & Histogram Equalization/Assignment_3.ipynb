```python
import cv2
import numpy as np
from math import acos, pi, sqrt, cos
```

```python
src = cv2.imread('test image/bridge.bmp', cv2.IMREAD_COLOR)
height, width, channel = src.shape
bgr = np.float32(src) / 255.0

b, g, r = cv2.split(bgr)

h = np.zeros((height, width), dtype=np.float)
s = np.zeros((height, width), dtype=np.float)
v = np.max(bgr, axis=2)

for i in range(height):
    for j in range(width):
        if v[i][j] == 0:
            h[i][j] = 0
            s[i][j] = 0
        else:
            min_rgb = min(bgr[i][j])

            s[i][j] = 1 - (min_rgb / v[i][j])

            if v[i][j] == r[i][j]:
                h[i][j] = 60 * (g[i][j] - b[i][j]) / (v[i][j] - min_rgb)
            elif v[i][j] == g[i][j]:
                h[i][j] = 120 + (60 * (b[i][j] - r[i][j])) / (v[i][j] - min_rgb)
            elif v[i][j] == b[i][j]:
                h[i][j] = 240 + (60 * (r[i][j] - g[i][j])) / (v[i][j] - min_rgb)
            if h[i][j] < 0:
                h[i][j] += 360
            h[i][j] /= 360

hsv_img = (np.dstack((h, s, v)) * 255).astype(np.uint8)

cv2.imshow('hsv', hsv_img)
cv2.imshow('h', h)
cv2.imshow('s', s)
cv2.imshow('v', v)

hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(hsv)
cv2.imshow('hsv2', hsv)
    
    
cv2.imshow("H", H)
cv2.imshow("S", S)
cv2.imshow("V", V)
#cv2.imshow("R", R)
#cv2.imshow("G", G)
#cv2.imshow("B", B)
#cv2.imshow("hsi2R", hsi2R)
#cv2.imshow("hsi2G", hsi2G)
#cv2.imshow("hsi2B", hsi2B)
#cv2.imshow("negative", negative)
#cv2.imshow("lambda 0.2", plt_1)
#cv2.imshow("lambda 0.4", plt_2)
#cv2.imshow("lambda 1.5", plt_3)
#cv2.imshow("lambda 5.0", plt_4)
cv2.imshow("src", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
